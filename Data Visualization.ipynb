# Alan Grafikleri

**Alan grafikleri** varsayılan oalrak yığınlanır. Yığılmamış bir alan grafiği oluşturmak için her sütun yada tüm pozitif uada negatif değerler olmalıdır. Yani herhangi bir NaN değeri varsayolan olarak "0" olacaktır. Sabitlenmiş bir grafik oluşturmak için "*stacked*" parametresini false vermeniz gerekmektedir. Ayrıca *figsize* parametresi bir tuple kabul eder ve oluşturduğunuz figürün boyutlarını belirtirsiniz. Geldik son ve en önemli parametreye "*kind*" bu parameter oluşturacağınız grafiğin tipini belirtir.

**Not:** NaN değeri bizim için veri tabanındaki "null" değerini buradaki karşılığıdır. Alan grafikleri boş değer kabul etmedikleri için dataframe'de boş olarak yanin NaN algıladıkları yerlere varsayılan oalrak sıfır doldurmaktadırlar.

**Not:** Yukarıda matplotlib kütüphanesini eklerken yanında pyplot argümanını kullandık. Bu argüman bize daha fazla özellik kazandırdır. Örneğin oluşturduğumuz grafiğe başlık eklemek (*title*), x ve y düzlemindeki verilerin ne anlama geldiğini paydaşlara daha iyi anlata bilmek adına kullandığımız *xlabel* ve *ylabel* gibi.

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(
    kind='area',
    stacked=False,
    figsize=(20,10)
)

plt.title('Immigration Trend of Top 5 Contries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

# Histogramlar

**Histogram** sayısal veri kümesinin frekans dağılımını temsil etmein bir yoludur. Çalışma şekli *x* eksenini bölmelere ayrılması, veri lümemizdeki her veri noktasını bir bölmeye ataması ve ardından her bölmeye atanmış veri noktası sayısını saymadıdır. Böylece y ekseni, her bölmedeki veri noktalarının frekansı veya sayısıdır diyebiliriz. Grafiğin boyutunu değiştirebileceğimizi ve dağılımın iyi görüntülenmesi  için genellikle ince ayar yapılması gerektğini unutmayın. Histogramlar en anlaşılık okuması paydaşlar tarafından daha kolay olan grafik türleridir.

```python
# 2013 yılında çeşitli ülkelerden Kanada'ya yeni göçmen sayısının nüfüs sıklığı nedir?
```

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])

print(count) # sıklığı-aralığı (frequency)
print(bin_edges) # bölme sayısı
# Varsayılan olarak histogram yöntemi veri kümesini 10 bölmeye ayırır.
```

```python
df_can['2013'].plot(
    kind='hist',
    figsize=(10,7),
    xticks = bin_edges,
    color= 'b'
)

plt.title('Histogram of Immigration from 195 Conuntries in 2013')
plt.ylabel('Number of Conuntries')
plt.xlabel('Number of Immigrants')
plt.grid(True)
plt.show()
```

# Çubuk Grafikler

Çubuk grafiği, çubukların uzunluğunun özellik / değişkenin büyüklüğünü / boyutunu temsil ettiği verileri temsil etmenin bir yoludur. Çubuk grafikler genellikle aralıklarla gruplandırılmış sayısal ve kategorik değişkenleri temsil eder.

**kind = bar** dikey bir çubuk grafiği <br>
**kind = barh** bir çubuk garfik oluşturur <br>

**Dikey çubuk çizim**

Dikey çubuk grafiklerde, x ekseni etiketleme için kullanılır ve y eksenindeki çubukların uzunluğu, ölçülen değişkenin büyüklüğüne karşılık gelir. Yatay çubuk grafiklerde, y ekseni etiketleme için kullanılır ve x ekseni üzerindeki çubukların uzunluğu, ölçülen değişkenin büyüklüğüne karşılık gelir. #### Pasta grafikler

Pasta grafik, bir daireyi (veya pasta) orantılı dilimlere bölerek sayısal oranları gösteren bir dairesel grafiktir. Büyük olasılıkla iş ve medyada yaygın olarak kullanıldığı için pasta grafiklerini zaten biliyorsunuzdur. Kind = pie anahtar sözcüğünü ileterek Matplotlib'de pasta grafikler oluşturabiliriz.

1980'den 2013'e kadar tüm dönem boyunca kıtalara göre gruplandırılmış yeni göçmenlerin oranını (yüzdesini) keşfetmek için bir pasta grafik kullanalım.

```python
# ülkeleri kıtalara göre gruplandırın ve sum () işlevini uyguluyoruz
df_continents = df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent', axis=0)))
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', # yüzdeleri ekliyoruz
                            startangle=90,     # başlangıç açısı 90° (Africa: başlangıç açısı başlangıç noktasını temsil eder)
                            shadow=True,       # gölgelendirme ekledik     
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Pasta grafiğini daire gibi görünecek şekilde ayanmasını sağlar

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # her kama için her kama dengelenmesi oranı.

df_continents['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # pasta grafikteki etiketleri kapatma
                            pctdistance=1.12,    # her pasta diliminin merkezi ile autopct tarafından üretilen metnin 
                                                 # başlangıcı arasındaki oran
                            colors=colors_list,  
                            explode=explode_list 
                            )

# başlığı pctdistance ile eşleştirmek için% 12 büyüt
plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12) 

plt.axis('equal') 

plt.legend(labels=df_continents.index, loc='upper left') 

plt.show()
```

#### Dağılım Grafikleri

Dağılım grafiği (2D), değişkenlerin birbirleriyle karşılaştırılmasında yararlı bir yöntemdir. Dağılım grafikleri, hem bağımsız hem de bağımlı değişkenleri 2B grafikte eşledikleri için çizgi grafiklerine benzer. Veri noktaları bir çizgi grafiğindeki bir çizgi ile birbirine bağlı olsa da, bir dağılım grafiğine bağlı değildir. Dağılım grafiğindeki verilerin bir eğilim ifade ettiği düşünülmektedir. Regresyon gibi araçları kullanarak daha fazla analiz yaparak, bu ilişkiyi matematiksel olarak hesaplayabilir ve veri kümesinin dışındaki eğilimleri tahmin etmek için kullanabiliriz.

```python
# yıllara göre toplam nüfusu elde etmek için sum () yöntemini kullanabiliriz
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

# Yılları int olarak değiştirin (bu grafiğe regrasyon uyarlayacağımız için buna ihtiyacımız olcak)
df_tot.index = map(int, df_tot.index)

# df_tot veri çerçevesinde bir sütun olarak geri koymak için dizini sıfırlayın
df_tot.reset_index(inplace = True)

# sütunu tekrardan isimledirdik
df_tot.columns = ['year', 'total']

# yaptığımız işlemlerin sonucu
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x = df_tot['year']      # x ekseni
y = df_tot['total']     # y ekseni
fit = np.polyfit(x, y, deg=1) 
# polyfit() => Yukarıda oluşturduğumuz veri çerçevelerine göre polinomunu hesaplar
fit
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# yıllara göre toplam göç miktarına en uygun fit polinomu plot üzerinde gösteriyoruz
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

#### Bubble Plots

A bubble plot is a variation of the scatter plot that displays three dimensions of data (x, y, z). The datapoints are replaced with bubbles, and the size of the bubble is determined by the third variable 'z', also known as the weight. In maplotlib, we can pass in an array or scalar to the keyword s to plot(), that contains the weight of each point.

```python
df_can_t = df_can[years].transpose() 

df_can_t.index = map(int, df_can_t.index)

df_can_t.index.name = 'Year'

df_can_t.reset_index(inplace=True)

df_can_t.head()
```

```python
# brazilya verisini normalize ediyoruz
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())

# Arjantin verisini normalize ediyoruz
norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

```python
# Brazil
ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,                 
                    color='green',
                    s=norm_brazil * 2000 + 10, 
                    xlim=(1975, 2015)
                   )

# Argentina
ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax = ax0
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
