## Eksik Değerleri Belirleme ve İşleme

#### Eksik Değerleri Belirleme

Veri setimizi incelediğinizde bazı sütunlar altında **?** olduğu gözünüze çarpmıştır. Bunun anlamı veri setimizde bazı eksik veriler olduğu anlamına gelir. Bu sembol yerine hesaplama hızı ve rahatlığı nedeniyle Python'ın varasyılan eksik değer işareti olan **NaN** ile veri setimizde bulunan soru işaretlerinin yerine Nan değerini basacağız.

Bu işlem için *Numpy* kütüphanesinede ihtiyacımız bulunmaktadır. Ayrıca *Pandas* kütüphanesine ait *replace()* metodu ile soru işaretlerinin yerine Nan değer basacağız.

*Documentation:* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html

```python
import numpy as np

# "?" yerine NaN değer basacağız
df.replace("?", np.nan, inplace = True)
df.head(5)

# replace() => foksiyonun içerisinde kullandığımız "inplace" argümanı ile yapılan değişişkliğin hemen 
# ilgili dataframe yansıtılmasını temin ettik
```

#### Eksik Verileri Değerlendirme

Eksik değerleri Python'ın varsayılan değerine yani *NaN* dönüştürülür. Bu eksik değerleri tanımlamak için Python'un yerleşik işlevlerini kullanıyoruz. 

Eksik verileri tespit etmek için iki yöntem vardır:
- isnull() => https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.isnull.html
- notnull() => https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.notnull.html

Her iki metodunda çıktısı, bağımsız değişkene iletilen değerin aslında eksik veri olup olmadığını gösteren bir boolean değeridir.

```python
missing_data = df.isnull()
missing_data.head(5)

# Aşağıdaki sonucu incelediğinizde anlamanız gereken nokta eksik değer içeren hücreler "True" değerini,
# eksik değer içermeyen hücreler ise "False" değeri almıştır.
```

#### Eksik Değerlerin Sayılması

Python'da bir for döngüsü kullanarak, her sütundaki eksik değerlerin sayısını hızlı bir şekilde bulabiliriz. Yukarıda belirtildiği gibi, "True" eksik bir değeri, "False" ise değerin veri kümesinde mevcut olduğu anlamına gelir. For döngüsünün gövdesinde ".value_counts ()" yöntemi "True" değerlerinin sayısını sayar.

*value_counts():* Pandas value_counts() işlevi, benzersiz değerlerin sayısını içeren nesneyi döndürür. Ortaya çıkan nesne, ilk elemanın en sık meydana gelen eleman olması için azalan sırada olacaktır. Varsayılan olarak NA değerlerini içermez.

*Documentation:* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

#### Eksik Verilerle Nasıl Başa Çılır?

Eksik veriler ile uğraşırken karşımıza temel iki yöntem ve onların alt yöntemleri çıkmaktadır. Bunlar
- Verileri Silme
    - Tüm bir satırı silme
    - Tüm bir sütunu silme
- Verileri Değiştirme
    - Ortalama değere göre değiştir
    - Frekans değerine göre değiştir
    
Tüm sütunlar yalnızca sütundaki girişlerin çoğu boşsa silinmelidir. Veri kümemizde hiçbir sütun tamamen silinmeyecek kadar boş değildir. Verilerin yerini alacak yöntemi seçme özgürlüğümüz vardır; ancak, bazı yöntemler diğerlerinden daha makul görünebilir. Her yöntemi birçok farklı sütuna uygulayacağız:

###### Ortalama Değer İle Değiştirin

Bu yöntemde ilgili sütunda tutulan değerlerin ortalaması alınır ve eksik veri içeren hücrelere bu ortalam değer basılır.

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# "normalized-losses" sütunun altındaki hücrelerde bulunan "NaN" değerlerine yukarıda elde ettiğimiz 
# ortalama değerle değiştiriyoruz
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df.head()
```

```python
# Belirli bir sütunda hangi değerlerin bulunduğunu görmek için ".value_counts ()" yöntemini kullanabiliriz:
# df["normalized-losses"].value_counts()

# Araçlardan dört kapının en yaygın tip olduğunu görebiliriz. Bizim için en yaygın türü otomatik olarak hesaplamak 
# için ".idxmax ()" yöntemini de kullanabiliriz:
df['num-of-doors'].value_counts().idxmax()
```

```python
# 'bore' sütununun ortalama değerini hesaplıyoruz
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
# NaN'yi ortalama değere göre değiştiriyor
df["bore"].replace(np.nan, avg_bore, inplace=True)

# 'horsepower' sütununun ortalama değerini hesaplıyoruz
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
# NaN'yi ortalama değere göre değiştiriyor
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

# 'peak-rpm' sütununun ortalama değerini hesaplıyoruz
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
# NaN'yi ortalama değere göre değiştiriyor
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

## Veri Standardizasyonu

Veri Standardizasyonu , ortalamayı çıkardığımız ve standart sapmaya böldüğümüz belirli bir veri normalizasyonu türü için bir terimdir.

###### Standardizasyon nedir?

Standardizasyon, verileri araştırmacının anlamlı bir karşılaştırma yapmasını sağlayan ortak bir formata dönüştürme işlemidir.

Misal

MPG'yi L / 100km'ye dönüştürün:

Veri setimizde, yakıt tüketimi sütunları "city-mpg" ve "highway-mpg" mpg (galon başına mil) birimi ile temsil edilmektedir. L / 100km standardında yakıt tüketimini kabul eden bir ülkede bir uygulama geliştirdiğimizi varsayalım

MPG'yi L / 100km'ye dönüştürmek için veri dönüşümünü uygulamamız gerekecek mi?

Birim dönüşümü için formül

L / 100km = 235 / mpg

Bu tarz işmeleri doğrudan Pandas ile yapabiliriz.

```python
df.head()
```

```python
# Matematiksel işlemle mpg'yi L / 100km'ye dönüştürüyoruz (235, mpg'ye bölünür)
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]

df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)

df.head()
```

## Veri Normalizasyonu

#### Neden normalleştirme?

Normalleştirme, birkaç değişkenin değerlerini benzer bir aralığa dönüştürme işlemidir. Tipik normalleştirmeler, değişkenin ortalama ortalaması 0 olacak şekilde ölçeklendirilmesini, değişkenin 1 olması için değişkenin ölçeklendirilmesini veya değişken değerlerin 0 ila 1 arasında değişmesini ölçeklendirmeyi içerir.

###### Örneğin

Normalleştirmeyi göstermek için "uzunluk", "genişlik" ve "yükseklik" sütunlarını ölçeklendirmek istediğimizi varsayalım

*Hedef:* Değerleri 0 ile 1 arasında değişecek şekilde bu değişkenleri Normalleştirmek istiyoruz.

*Yaklaşım:* orijinal değeri (orijinal değer) / (maksimum değer) ile değiştirin

```python
# (orijinal değer) yerine (orijinal değer) / (maksimum değer)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
# "height" sütununu normalleştirelim
df['height'] = df['height']/df['height'].max() 
df[["length","width","height"]].head()
```

## Binning (Kutulama)

Neden binning?

Binning, gruplandırılmış analiz için sürekli sayısal değişkenleri ayrık kategorik 'kutulara' dönüştürme sürecidir.

**Örneğin:** Veri setimizde, "beygir gücü" 48 ile 288 arasında değişen gerçek değerli bir değişkendir, 57 benzersiz değere sahiptir. Ya sadece yüksek beygir gücü, orta beygir gücü ve az beygir gücü (3 tip) olan araçlar arasındaki fiyat farkını önemsersek? Analizi basitleştirmek için bunları üç bölmeye yeniden yerleştirebiliriz. *'Beygir gücü'* sütununu 3 bölmeye ayırmak için *'kes'* panda yöntemini kullanacağız

```python
# Verileri doğru biçime dönüştürme
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

Eşit büyüklükte 3 bant genişliği istiyoruz, bu nedenle numpy'nin linspace() metodunu kullanacağız.

*Documentation:* https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

Minimum beygir gücü değerini dahil etmek istediğimizden start_value = min (df ["beygir gücü"]) değerini ayarlamak istiyoruz.

Maksimum beygir gücü değerini dahil etmek istediğimiz için end_value = max (df ["beygir gücü"]) değerini ayarlamak istiyoruz.

Eşit uzunlukta 3 kutu inşa ettiğimiz için 4 bölücü olmalı, bu nedenle numbers_generated = 4 oalrak belirtiyoruz.

Yukarıda hesaplanan bant genişliği ile minimum değeri maksimum değere sahip bir bin dizisi oluşturuyoruz. Kutular, bir kutunun ne zaman biteceğini ve diğerinin ne zaman başlayacağını belirlemek için kullanılan değerler olacaktır.

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
# Her "df ['beygir gücü']" değerinin neye ait olduğunu belirlemek için "cut" işlevini uyguluyoruz.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head()
```

```python
# Her bölmedeki araç sayısını görelim.
df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

## Dummy Variable or Indicator variable 

Bir gösterge değişkeni (veya kukla değişken), kategorileri etiketlemek için kullanılan sayısal bir değişkendir. Bunlara 'aptallar' denir, çünkü sayıların kendilerinin doğal bir anlamı yoktur.

**Örneğin

"Yakıt tipi" sütununda "gaz" veya "dizel" olmak üzere iki benzersiz değer vardır. Regrasyon algoritmaları kelimeleri ile çalışamaz, sadece rakamları algıalyabilir ve kullanabilir. Regresyon analizinde bu özelliği kullanmak için, "yakıt tipi" ni gösterge değişkenlerine dönüştürüyoruz.

Farklı yakıt türü kategorilerine sayısal değerler atamak için panda'nın 'get_dummies()' yöntemini kullanacağız.

*Documentation:* https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html

```python
# gösterge değişkenlerini alın ve bunu "dummy_variable_1" veri çerçevesine atayalım
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
# anlaşılır olması için sütun adlarını değiştirme
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# "df" ve "dummy_variable_1" veri çerçevesini birleştir
df = pd.concat([df, dummy_variable_1], axis=1)

# "df" den orijinal sütun "yakıt türü" siliyoruz
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df.head()
```

```python
# "aspirasyon" sütununa gösterge değişkeni yaratalım: "std" ila 0, "turbo" ila 1.

# aspirasyon değişkenlerini al ve veri çerçevesine "dummy_variable_2" ata
dummy_variable_2 = pd.get_dummies(df['aspiration'])

# anlaşılır olması için sütun adlarını değiştirme
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)

# "dummy_variable_1" veri çerçevesinin ilk 5 örneğini gösterelim
dummy_variable_2.head()
```
